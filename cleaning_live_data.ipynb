```python
df = data.dropna(axis=0, subset=['ServiceJson'])
```

```python
data1=df.iloc[:,[1,7]]
```

```python
data1
```

```python
import yaml
complaints=[]
for i in data1['ServiceJson']:

    complaints.append(yaml.load(i,Loader=yaml.FullLoader))
```

```python
questions=[]
for i in complaints:
    for j in i:
        if type(j)==dict:
            questions.extend(j['ChatBotMsg'])
    questions.append("hi")       
```

```python
service_name=[]
for i in data1['ServiceName']:
    service_name.append(i)
```

```python

from itertools import groupby

# define separator keys
def split_condition(x):
    return x in {'hi'}

# define groupby object
grouper = groupby(questions, key=split_condition)

# convert to dictionary via enumerate
res = dict(enumerate((list(j) for i, j in grouper if not i), 1))
```

```python
df2 = pd.DataFrame.from_dict(res, orient ='index') 
```

```python
df2['patterns']=service_name
```

```python
first_column = df2.pop('patterns')
  
# insert column using insert(position,column_name,
# first_column) function
df2.insert(0, 'patterns', first_column)
```

```python
df2.fillna(value="",inplace=True)
```
